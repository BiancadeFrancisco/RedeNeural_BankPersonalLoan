```python
import pandas as pd
```

```python
df = pd.read_excel("/content/Bank_Personal_Loan_Modelling.xlsx")

df.head()
```

```python
# NESSE CASO A COLUNA ID NÃO É NECESSÁRIA, PODEMOS EXCLUI-LA:

df.drop(columns=["ID"])
```

```python
df.describe()
```

```python
# VERIFICAR A DISTRIBUIÇÃO DA NOSSA VARIÁVEL TARGET:

df["Personal_Loan"].value_counts() # resultado: não 4520, sim 480
```

```python
# SEPARAR AS VARIÁVEIS:

x = df.drop(columns=["Personal_Loan"])  # todas, menos a coluna Personal_Loan (target)

y = df["Personal_Loan"]
```

```python
# SEPARAR TREINO E TESTE:

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test, = train_test_split(x, y, test_size=0.2)
```

```python
# FAZER NORMALIZAÇÃO DAS NOSSAS VARIÁVEIS:

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
# TRANSFORMAR TUDO EM DATAFRAME:

X_train = pd.DataFrame(X_train_scaled, columns = X_train.columns)

X_test = pd.DataFrame(X_test_scaled, columns = X_test.columns)
```

```python
# IMPLEMENTAR A REDE NEURAL:

from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

clf = MLPClassifier(hidden_layer_sizes = (100, 50, 20), activation = "logistic")
# hiperparâmetro (camadas escondidas = "hidden_layer_sizes = (100, 50, 20)") e a função de ativação (sigmoide = "activation = "logistic")

clf.fit(X_train, y_train)
```

```python
# VERIFICAR A ACURÁCIA E O F1 DO MODELO:

y_pred = clf.predict(X_test)
print(f"Acurácia: {accuracy_score(y_test, y_pred)*100}%")
print(f"F1 Score: {f1_score(y_test, y_pred)*100}%")
```
